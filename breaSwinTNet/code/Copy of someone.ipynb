```python
# !pip install tensorflow
# !pip install opencv-python
import cv2
import tensorflow as tf
import os
import sys
sys.path.append('./Swin-Transformer-TF')
from swintransformer import SwinTransformer
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import RepeatedStratifiedKFold, train_test_split
from vit_keras import vit, utils, visualize
```

```python
data_dir = '/home/afrida/Documents/skin_cancer_classification/raw_data/class_separated_data'
akiec = os.path.join(data_dir, 'akiec')
bcc = os.path.join(data_dir, 'bcc')
bkl = os.path.join(data_dir, 'bkl')
df = os.path.join(data_dir, 'df')
mel = os.path.join(data_dir, 'mel')
nv = os.path.join(data_dir, 'nv')
vasc = os.path.join(data_dir, 'vasc')
```

```python
def extract_images(path, sizex, sizey):
  data = []
  folders = sorted(os.listdir(path))
  for file in folders:
    mat_file = cv2.imread(os.path.join(path, file))
    data.append(cv2.resize(mat_file, (sizex, sizey)))
    print(np.array(data).shape)
  return np.array(data)
```

```python
sizex, sizey = 224,224
akiec_mat = extract_images(akiec, sizex, sizey)
bcc_mat = extract_images(bcc, sizex, sizey)
bkl_mat = extract_images(bkl, sizex, sizey)
df_mat = extract_images(df, sizex, sizey)
mel_mat = extract_images(mel, sizex, sizey)
nv_mat = extract_images(nv, sizex, sizey)
vasc_mat = extract_images(vasc, sizex, sizey)
```

```python
np.save(os.path.join('./akiec_mat.npy'), akiec_mat)
np.save(os.path.join('./bcc_mat.npy'), bcc_mat)
np.save(os.path.join('./bkl_mat.npy'), bkl_mat)
np.save(os.path.join('./df_mat.npy'), df_mat)
np.save(os.path.join('./mel_mat.npy'), mel_mat)
np.save(os.path.join('./nv_mat.npy'), nv_mat)
np.save(os.path.join('./vasc_mat.npy'), vasc_mat)
```

```python
import numpy as np
import os
akiec_mat= np.load(os.path.join('./akiec_mat.npy'))
bcc_mat= np.load(os.path.join('./bcc_mat.npy'))
bkl_mat= np.load(os.path.join('./bkl_mat.npy'))
df_mat= np.load(os.path.join('./df_mat.npy'))
mel_mat= np.load(os.path.join('./mel_mat.npy'))
nv_mat= np.load(os.path.join('./nv_mat.npy'))
vasc_mat= np.load(os.path.join('./vasc_mat.npy'))
```

```python
akiec_labels = np.zeros(akiec_mat.shape[0])
bcc_labels = np.ones(bcc_mat.shape[0])
bkl_labels = 2*np.ones(bkl_mat.shape[0])
df_labels = 3*np.ones(df_mat.shape[0])
mel_labels = 4*np.ones(mel_mat.shape[0])
nv_labels = 5*np.ones(nv_mat.shape[0])
vasc_labels = 6*np.ones(vasc_mat.shape[0])
```

```python
X = np.concatenate((akiec_mat,bcc_mat,bkl_mat,df_mat,mel_mat,nv_mat,vasc_mat), axis = 0)
```

```python
y = np.concatenate((akiec_labels,bcc_labels,bkl_labels,df_labels,mel_labels,nv_labels,vasc_labels), axis=0)
```

```python
X = vit.preprocess_inputs(X)
print(np.min(X))
print(np.max(X))
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.4)
```

```python
import numpy as np
np.save("./x_train.npy",x_train)
np.save("./y_train.npy",y_train)
np.save("./x_test.npy",x_test)
np.save("./y_test.npy",y_test)
```

```python
x_train = np.load("./x_train.npy")
y_train=np.load("./y_train.npy")
```

```python
IMAGE_SIZE = [x_train.shape[1], x_train.shape[1], 3]
NUM_CLASSES = 7

swin_vit = SwinTransformer('swin_tiny_224', include_top=False, pretrained=True)

model = tf.keras.Sequential()
model.add(tf.keras.Input(shape = IMAGE_SIZE))
model.add(swin_vit)
model.add(tf.keras.layers.Dense(NUM_CLASSES, activation='softmax'))

model.build(IMAGE_SIZE)
```

```python
model.summary()

model.compile(optimizer = tf.keras.optimizers.Adadelta(learning_rate = 0.1), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, batch_size = 4, epochs = 15, validation_split = 0.10)
```

```python
x_test = np.load("./x_test.npy")
y_test=np.load("./y_test.npy")
```

```python
model.evaluate(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
y_prob = np.argmax(y_pred, axis = 1)
```

```python
import sklearn

print(f'Accuracy {sklearn.metrics.accuracy_score(y_test, y_prob)}')
AUC = sklearn.metrics.roc_auc_score(y_test, y_pred, multi_class='ovo')
print(f'AUC is {AUC}')
target_names = ["akiec", "bcc", "bkl", "df", "mel", "vasc", "nv"]
print('F1-score', sklearn.metrics.f1_score(y_test, y_prob, average = 'micro'))
print(f'Balanced accuracy {sklearn.metrics.balanced_accuracy_score(y_test, y_prob)}')
print(f'MCC {sklearn.metrics.matthews_corrcoef(y_test, y_prob)}')
print(f'{sklearn.metrics.classification_report(y_test, y_prob, target_names=target_names)}')
```
